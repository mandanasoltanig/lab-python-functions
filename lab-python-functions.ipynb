```python
products = ["t-shirts", "mug", "hat", "book", "keychain"]
def initialize_inventory(products):
    inventory = {}

    for product in products:
        quantity = int(input(f"Enter quantity for {product}: "))
        inventory[product] = quantity

    return inventory
```

```python
def get_customer_orders():
    customer_orders = set()

    while True:
        product = input("Enter a product you need to order, (or type 'done' to finish): ")
        if product == "done":
           break

        customer_orders.add(product)

    return customer_orders
```

```python
def update_inventory(customer_orders, inventory):
    for product in customer_orders:
        if product in inventory:
            inventory[product] -= 1
    return inventory
            
```

```python
def calculate_order_statistics(customer_orders, products):
    total_products_ordered = len(customer_orders)

    percentage_unique= (total_products_ordered / len(products)) * 100

    order_statistics = {
          "total_products_ordered": total_products_ordered,
          "percentage_unique": percentage_unique
    }

    return order_statistics
```
